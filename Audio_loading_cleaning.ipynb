```python
import librosa
import os
import numpy as np
from IPython.display import Audio

def load_audio_files(folder_path, sample_rate=22050, max_files=2000):
    audio_data = []
    file_count = 0

    for file in os.listdir(folder_path):
        if file_count >= max_files:
            break

        file_path = os.path.join(folder_path, file)
        if os.path.isfile(file_path):
            try:
                data, _ = librosa.load(file_path, sr=sample_rate)
                audio_data.append(data)
                file_count += 1
            except Exception as e:
                print(f"Error loading {file_path}: {e}")

    return audio_data

# Paths to the folders containing clean and noisy audio files
clean_audio_folder = 'clean_audio/'  # Replace with the path to Mozilla Common Voice dataset
noisy_audio_folder = 'noise_audio/'  # Replace with the path to Urban8K dataset

# Load the audio files
clean_audio_data = load_audio_files(clean_audio_folder)
noisy_audio_data = load_audio_files(noisy_audio_folder)

# Example: Accessing the first clean and noisy audio data
first_clean_audio = clean_audio_data[0]
display(Audio(first_clean_audio, rate=22050))

first_noisy_audio = noisy_audio_data[0]
display(Audio(first_noisy_audio, rate=22050))
```

```python
import librosa
import os
import soundfile as sf

def process_and_save_audio_files(folder_path, output_folder, sample_rate=22050, max_duration=10):
    for file in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file)
        if os.path.isfile(file_path):
            try:
                data, sr = librosa.load(file_path, sr=sample_rate)
                
                # Trim silent frames
                non_silent_data, _ = librosa.effects.trim(data)

                # Check duration for clean audio
                duration = librosa.get_duration(y=non_silent_data, sr=sr)
                if duration <= max_duration:
                    output_path = os.path.join(output_folder, file)
                    sf.write(output_path, non_silent_data, sr)
            except Exception as e:
                print(f"Error processing {file_path}: {e}")

# Paths to the folders containing clean and noisy audio files
clean_audio_folder = 'clean_audio/'  # Replace with the path to Mozilla Common Voice dataset
noisy_audio_folder = 'noise_audio/'  # Replace with the path to Urban8K dataset

# Output folders for processed audio files
processed_clean_folder = 'clean_processed/'  # Replace with your output path
processed_noisy_folder = 'noisy_processed/'  # Replace with your output path

# Process and save audio files
process_and_save_audio_files(clean_audio_folder, processed_clean_folder)
process_and_save_audio_files(noisy_audio_folder, processed_noisy_folder)
```

```python
import librosa
import os
import soundfile as sf
import numpy as np
import random

def adjust_and_add_noise(clean_data, noise_data):
    # Adjust the length of the noise data to match the clean data
    if len(noise_data) < len(clean_data):
        # Repeat the noise data
        repeat_times = int(np.ceil(len(clean_data) / len(noise_data)))
        noise_data = np.tile(noise_data, repeat_times)[:len(clean_data)]
    else:
        # Truncate the noise data
        noise_data = noise_data[:len(clean_data)]

    # Add the noise to the clean data
    noisy_data = clean_data + noise_data
    return noisy_data

def create_noisy_dataset(clean_folder, noise_folder, output_folder, sample_rate=22050):
    noise_files = os.listdir(noise_folder)

    for clean_file in os.listdir(clean_folder):
        clean_path = os.path.join(clean_folder, clean_file)

        # Randomly select a noise file
        noise_file = random.choice(noise_files)
        noise_path = os.path.join(noise_folder, noise_file)

        try:
            clean_data, _ = librosa.load(clean_path, sr=sample_rate)
            noise_data, _ = librosa.load(noise_path, sr=sample_rate)

            # Create noisy audio by adding noise to clean audio
            noisy_data = adjust_and_add_noise(clean_data, noise_data)

            # Save the noisy audio
            output_file_name = os.path.splitext(clean_file)[0] + '_noisy.wav'
            sf.write(os.path.join(output_folder, output_file_name), noisy_data, sample_rate)
        except Exception as e:
            print(f"Error processing file {clean_file}: {e}")
#folders
clean_folder = 'clean_preprocessedwav/'  # Replace with your output path
noisy_folder = 'noisy_processed/'
output_noisy_dataset_folder = 'merged_noisy/'

# Create and save the noisy dataset
create_noisy_dataset(clean_folder, noisy_folder, output_noisy_dataset_folder)
```

```python
import librosa
import numpy as np
import librosa.display
import matplotlib.pyplot as plt
from scipy.ndimage import zoom

def compute_spectrogram(audio_file, n_fft=2048, window_length=255, fft_length=255, hop_length=63, size=128):
    # Load the audio file
    audio, sr = librosa.load(audio_file, sr=None)

    # Compute the Short Term Fourier Transform (STFT)
    stft = librosa.stft(audio, n_fft=n_fft, hop_length=hop_length)

    # Compute the magnitude (power) spectrogram
    spectrogram = np.abs(stft)**2

    # Convert to log scale (dB)
    log_spectrogram = librosa.amplitude_to_db(spectrogram, ref=np.max)

    # Resize to 128x128
    spectrogram_resized = zoom(log_spectrogram, (size / log_spectrogram.shape[0], size / log_spectrogram.shape[1]))

    return spectrogram_resized

# Example usage
spectrogram = compute_spectrogram('merged_noisy/common_voice_en_37285758_noisy.wav')

spec = compute_spectrogram('noisy_processed/518-4-0-2.wav')
spec2 = compute_spectrogram('clean_preprocessedwav/common_voice_en_37285758.wav')

 # Plot
plt.figure(figsize=(10, 8))
librosa.display.specshow(spectrogram, sr=22050, hop_length=125, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram')
plt.show()

 # Plot
plt.figure(figsize=(10, 8))
librosa.display.specshow(spec, sr=22050, hop_length=125, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram')
plt.show()

 # Plot
plt.figure(figsize=(10, 8))
librosa.display.specshow(spec2, sr=22050, hop_length=125, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram')
plt.show()
```

```python
import librosa
import numpy as np
import os
from scipy.ndimage import zoom

def compute_and_save_spectrogram_as_npy(audio_file, output_file, window_length, fft_length, hop_length, size=128):
    # Load the audio file
    audio, sr = librosa.load(audio_file, sr=None)

    # Apply windowing and padding
    audio_padded = librosa.util.fix_length(audio, size=len(audio) + window_length // 2)

    # Compute the Short Term Fourier Transform (STFT)
    stft = librosa.stft(audio_padded, n_fft=fft_length, hop_length=hop_length, win_length=window_length)

    # Compute the magnitude (power) spectrogram
    spectrogram = np.abs(stft)**2

    # Convert to log scale (dB)
    log_spectrogram = librosa.amplitude_to_db(spectrogram, ref=np.max)

    # Resize the spectrogram to 128x128
    spectrogram_resized = zoom(log_spectrogram, (size / log_spectrogram.shape[0], size / log_spectrogram.shape[1]))

    # Save the resized spectrogram as a numpy array
    np.save(output_file, spectrogram_resized)

def process_folder(input_folder, output_folder, window_length, fft_length, hop_length, size=128):
    for file in os.listdir(input_folder):
        if file.endswith(".wav"):
            input_file_path = os.path.join(input_folder, file)
            output_file_path = os.path.join(output_folder, os.path.splitext(file)[0] + ".npy")
            compute_and_save_spectrogram_as_npy(input_file_path, output_file_path, window_length, fft_length, hop_length, size)

# Define your input and output folder paths
noisy_folder = 'merged_noisy/'
clean_folder = 'clean_preprocessedwav/'
output_noisy = 'merged_spectrogram_npy/'
output_clean = 'clean_spectrogram_npy/'

# Define STFT parameters
windowLength = 255
fftLength = 255
hop_length = 63

# Process all audio files in the folder
process_folder(noisy_folder, output_noisy, windowLength, fftLength, hop_length)


process_folder(clean_folder, output_clean, windowLength, fftLength, hop_length)

```

```python
import numpy as np
import os

def load_spectrograms_from_directory(directory):
    spectrograms = []
    for file in os.listdir(directory):
        if file.endswith('.npy'):
            file_path = os.path.join(directory, file)
            spectrogram = np.load(file_path)
            # Check if the shape is as expected
            if spectrogram.shape != (128, 128):
                print(f"Unexpected shape for file {file}: {spectrogram.shape}")
            spectrograms.append(spectrogram)
    return np.array(spectrograms)


# Paths to the output directories
clean_spectrogram_dir = 'merged_spectrogram_npy/'
noisy_spectrogram_dir = 'clean_spectrogram_npy/'

# Load spectrogram data
clean_spectrograms = load_spectrograms_from_directory(clean_spectrogram_dir)
noisy_spectrograms = load_spectrograms_from_directory(noisy_spectrogram_dir)
```

```python
## Normalize the data
def normalize(data):
    data_norm = np.zeros_like(data)
    for i in range(len(data)):
        min_val = np.amin(data[i])
        max_val = np.amax(data[i])
        if max_val != min_val:
            data_norm[i] = (data[i] - min_val) / (max_val - min_val)
        else:
            # Handle the case where all values in data[i] are the same
            # For example, set data_norm[i] to zero or maintain as is
            data_norm[i] = data[i]  # Or set to zero if appropriate
    return data_norm
```

```python
#IF issues with GPU remove tensorflow.keras.* and only keep keras.*
from keras.models import Model
from keras.layers import Input, Conv2D, LeakyReLU, MaxPooling2D, Dropout, concatenate, UpSampling2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import backend
import tensorflow as tf

size_filter_in = 16
#normal initialization of weights
kernel_init = 'he_normal'
#To apply leaky relu after the conv layer 
activation_layer = None
inputs = keras.Input(shape=(128, 128, 1))
conv1 = Conv2D(size_filter_in, 3, activation = activation_layer, padding = 'same', kernel_initializer = kernel_init)(inputs)
conv1 = LeakyReLU()(conv1)
conv1 = Conv2D(size_filter_in, 3, activation = activation_layer, padding = 'same', kernel_initializer = kernel_init)(conv1)
conv1 = LeakyReLU()(conv1)
pool1 = MaxPooling2D(pool_size=(2, 2))(conv1)
conv2 = Conv2D(size_filter_in*2, 3, activation = activation_layer, padding = 'same', kernel_initializer = kernel_init)(pool1)
conv2 = LeakyReLU()(conv2)
conv2 = Conv2D(size_filter_in*2, 3, activation = activation_layer, padding = 'same', kernel_initializer = kernel_init)(conv2)
conv2 = LeakyReLU()(conv2)
pool2 = MaxPooling2D(pool_size=(2, 2))(conv2)
conv3 = Conv2D(size_filter_in*4, 3, activation = activation_layer, padding = 'same', kernel_initializer = kernel_init)(pool2)
conv3 = LeakyReLU()(conv3)
conv3 = Conv2D(size_filter_in*4, 3, activation = activation_layer, padding = 'same', kernel_initializer = kernel_init)(conv3)
conv3 = LeakyReLU()(conv3)
pool3 = MaxPooling2D(pool_size=(2, 2))(conv3)
conv4 = Conv2D(size_filter_in*8, 3, activation = activation_layer, padding = 'same', kernel_initializer = kernel_init)(pool3)
conv4 = LeakyReLU()(conv4)
conv4 = Conv2D(size_filter_in*8, 3, activation = activation_layer, padding = 'same', kernel_initializer = kernel_init)(conv4)
conv4 = LeakyReLU()(conv4)
drop4 = Dropout(0.5)(conv4)
pool4 = MaxPooling2D(pool_size=(2, 2))(drop4)

conv5 = Conv2D(size_filter_in*16, 3, activation = activation_layer, padding = 'same', kernel_initializer = kernel_init)(pool4)
conv5 = LeakyReLU()(conv5)
conv5 = Conv2D(size_filter_in*16, 3, activation = activation_layer, padding = 'same', kernel_initializer = kernel_init)(conv5)
conv5 = LeakyReLU()(conv5)
drop5 = Dropout(0.5)(conv5)

up6 = Conv2D(size_filter_in*8, 2, activation = activation_layer, padding = 'same', kernel_initializer = kernel_init)(UpSampling2D(size = (2,2))(drop5))
up6 = LeakyReLU()(up6)
merge6 = concatenate([drop4,up6], axis = 3)
conv6 = Conv2D(size_filter_in*8, 3, activation = activation_layer, padding = 'same', kernel_initializer = kernel_init)(merge6)
conv6 = LeakyReLU()(conv6)
conv6 = Conv2D(size_filter_in*8, 3, activation = activation_layer, padding = 'same', kernel_initializer = kernel_init)(conv6)
conv6 = LeakyReLU()(conv6)
up7 = Conv2D(size_filter_in*4, 2, activation = activation_layer, padding = 'same', kernel_initializer = kernel_init)(UpSampling2D(size = (2,2))(conv6))
up7 = LeakyReLU()(up7)
merge7 = concatenate([conv3,up7], axis = 3)
conv7 = Conv2D(size_filter_in*4, 3, activation = activation_layer, padding = 'same', kernel_initializer = kernel_init)(merge7)
conv7 = LeakyReLU()(conv7)
conv7 = Conv2D(size_filter_in*4, 3, activation = activation_layer, padding = 'same', kernel_initializer = kernel_init)(conv7)
conv7 = LeakyReLU()(conv7)
up8 = Conv2D(size_filter_in*2, 2, activation = activation_layer, padding = 'same', kernel_initializer = kernel_init)(UpSampling2D(size = (2,2))(conv7))
up8 = LeakyReLU()(up8)
merge8 = concatenate([conv2,up8], axis = 3)
conv8 = Conv2D(size_filter_in*2, 3, activation = activation_layer, padding = 'same', kernel_initializer = kernel_init)(merge8)
conv8 = LeakyReLU()(conv8)
conv8 = Conv2D(size_filter_in*2, 3, activation = activation_layer, padding = 'same', kernel_initializer = kernel_init)(conv8)
conv8 = LeakyReLU()(conv8)

up9 = Conv2D(size_filter_in, 2, activation = activation_layer, padding = 'same', kernel_initializer = kernel_init)(UpSampling2D(size = (2,2))(conv8))
up9 = LeakyReLU()(up9)
merge9 = concatenate([conv1,up9], axis = 3)
conv9 = Conv2D(size_filter_in, 3, activation = activation_layer, padding = 'same', kernel_initializer = kernel_init)(merge9)
conv9 = LeakyReLU()(conv9)
conv9 = Conv2D(size_filter_in, 3, activation = activation_layer, padding = 'same', kernel_initializer = kernel_init)(conv9)
conv9 = LeakyReLU()(conv9)
conv9 = Conv2D(2, 3, activation = activation_layer, padding = 'same', kernel_initializer = kernel_init)(conv9)
conv9 = LeakyReLU()(conv9)
conv10 = Conv2D(1, 1, activation = 'tanh')(conv9)

model = Model(inputs,conv10)

model.compile(optimizer = 'adam', loss = tf.keras.losses.Huber(), metrics = ['mae'])
#model.compile(optimizer = 'adam', loss = tf.keras.losses.Huber(), metrics = ['mae'])

model.summary()
```

```python
def read_audio(filepath, sample_rate, normalize=True):
    # print(f"Reading: {filepath}").
    
    warnings.filterwarnings("ignore")
    warnings.warn("PySoundFile failed. Trying audioread instead.")
    audio, sr = librosa.load(filepath, sr=sample_rate)
    if normalize:
      div_fac = 1 / np.max(np.abs(audio)) / 3.0
      audio = audio * div_fac
    return audio, sr
        
def add_noise_to_clean_audio(clean_audio, noise_signal):
    if len(clean_audio) >= len(noise_signal):
#         print("The noisy signal is smaller than the clean audio input. Duplicating the noise.")
        while len(clean_audio) >= len(noise_signal):
            noise_signal = np.append(noise_signal, noise_signal)

    ## Extract a noise segment from a random location in the noise file
#     ind = np.random.randint(0, noise_signal.size - clean_audio.size)

#     noiseSegment = noise_signal[ind: ind + clean_audio.size]

    noiseSegment = noise_signal[0:63875]
        

    speech_power = np.sum(clean_audio ** 2)
    noise_power = np.sum(noiseSegment ** 2)
    noisyAudio = clean_audio + np.sqrt(speech_power / noise_power) * noiseSegment
#     print(len(clean_audio),len(noisyAudio))
    return noisyAudio

def play(audio, sample_rate):
    ipd.display(ipd.Audio(data=audio, rate=sample_rate))  # load a local WAV file


def truncate_and_resize_data(clean):
    start_point=0
    clean_abs = np.abs(clean)
    for i in range(len(clean_abs)):
        if(clean_abs[i]>0):
            start_point=i
            break
    if(len(clean)-start_point>63875):
        clean = clean[start_point:start_point+63875]
    else:
        clean = clean[start_point:len(clean)]
        clean = np.pad(clean,(0,63875-len(clean)),'constant')
    
    #print(len(noisy),len(clean))
        
    return clean


import numpy as np
import librosa
import scipy.signal
import scipy.ndimage

def get_spectogram(noisy, clean):
    # Set the FFT length and window length for 128 frequency bins
    n_fft = 128
    windowLength = n_fft

    # Calculate the hop length to get close to 128 time frames
    target_time_frames = 128
    hop_length = max(1, int(len(noisy) / target_time_frames))

    # Create the window
    window = scipy.signal.windows.hamming(windowLength, sym=False)

    # Compute the STFT
    spec_clean = librosa.stft(clean, n_fft=n_fft, win_length=windowLength, hop_length=hop_length,
                              window=window, center=True)
    S_db_clean = librosa.amplitude_to_db(np.abs(spec_clean), ref=np.max)

    spec_noisy = librosa.stft(noisy, n_fft=n_fft, win_length=windowLength, hop_length=hop_length,
                              window=window, center=True)
    S_db_noisy = librosa.amplitude_to_db(np.abs(spec_noisy), ref=np.max)
    ang_noisy = np.angle(spec_noisy)

    # Resize or interpolate the spectrogram and angle to get 128x128 dimensions
    S_db_clean_resized = scipy.ndimage.zoom(S_db_clean, (128 / S_db_clean.shape[0], 128 / S_db_clean.shape[1]))
    S_db_noisy_resized = scipy.ndimage.zoom(S_db_noisy, (128 / S_db_noisy.shape[0], 128 / S_db_noisy.shape[1]))
    ang_noisy_resized = scipy.ndimage.zoom(ang_noisy, (128 / ang_noisy.shape[0], 128 / ang_noisy.shape[1]))

    return S_db_noisy_resized, S_db_clean_resized, ang_noisy_resized
```

```python


def normalize_spec(spec):
  spec_max = np.amax(spec)
  spec_min = np.amin(spec)
  spec_norm = (spec-spec_min)/(spec_max-spec_min)
  return spec_norm,spec_max,spec_min

def de_norm_spec(spec,spec_max,spec_min):
  spec_denorm = (spec*(spec_max-spec_min))+spec_min
  return spec_denorm
     
```
